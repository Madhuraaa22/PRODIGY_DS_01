```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)
print(df.head())

df = df[['Country Name', '2020']]

df = df.dropna()

#Bar Chart
top10 = df.sort_values(by='2020', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(top10['Country Name'], top10['2020'] / 1e6)
plt.xticks(rotation=45)
plt.title("Top 10 Countries by Population in 2020")
plt.xlabel("Country")
plt.ylabel("population (millions)")
plt.tight_layout()
plt.show()

#Histogarm
plt.figure(figsize=(10,6))
plt.hist(df['2020'] / 1e6, bins=20, color='purple', edgecolor='black')
plt.title("Distribution of Country Population in 2020")
plt.xlabel("Population (millions)")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()
```
